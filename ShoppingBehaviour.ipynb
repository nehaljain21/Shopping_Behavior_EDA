```python
highest_revenue_category = df.groupby("Category")["Purchase Amount (USD)"].sum()
print(highest_revenue_category)
plt.figure(figsize=(7,5))
highest_revenue_category.plot(kind='pie',autopct='%1.1f%%')
plt.title('Total revenue by each category')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="Gender", hue="Category")
plt.title("Category distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
print (df["Subscription Status"].value_counts())
df["Subscription Status"].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Subscription Status of customers')
plt.xlabel('Subscription Status')
plt.ylabel('No. of customers')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='Subscription Status', hue='Category')
plt.title('Category Distribution by Subscription Status')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Frequency of Purchases', y='Purchase Amount (USD)')
plt.title('Purchase Amount by Frequency of Purchases')
plt.show()
```

```python
plt.figure(figsize=(7,5))
payment_data = df.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
print(payment_data)
payment_data.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Average Purchase Amount by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```

```python
avg_rating = df.groupby('Discount Applied')['Review Rating'].mean()
print(avg_rating)
plt.figure(figsize=(7, 5))
avg_rating.plot(kind="bar", edgecolor='black')
plt.title('Average Review Rating by Discount Applied')
plt.xlabel('Discount Applied')
plt.ylabel('Average Review Rating')
plt.show()
```

**Insights**

The dataset consists of **3,900** customer transactions, capturing both demographic and purchase-related attributes across **18** key features- including Age, Gender, Item Purchased, Category, Purchase Amount, Location, Size, Color, Season, Review Rating, Subscription Status, Shipping Type, Discount Applied, Payment Method, and Frequency of Purchases.
The data provides valuable insights into customer behavior, purchasing preferences, and product performance.

Key Observations from **Exploratory Data Analysis (EDA)**:

- Gender Distribution: Out of 3,900 customers, 68% are male and 32% are female.
Purchases are dominated by males across all product categories, suggesting that male customers form the majority in this dataset*(depicted in pie plot)*.

- Age Distribution: Customer ages range from 18 to 70 years, with frequencies distributed almost uniformly across the range*(as shown in histogram)*.

- Items Purchased:There are 25 distinct items in the dataset, with the top-purchased items being Blouse, Pants, Jewelry, Shirt, and Dress, each having over 165 purchases.Jeans and Gloves have the lowest purchase counts*(depicted in bar plot)*.

- Category-Wise Distribution:There are four primary product categories — Clothing, Accessories, Footwear, and Outerwear.Clothing accounts for the highest number of purchases and contributes the most to revenue (44.7%) followed by Accessories *(visualized in pie chart)*. Both genders predominantly purchase Clothing, but males show higher purchase frequencies across all categories.*(as visualised in countplot)*

- Size Distribution:There are four size variants — S, M, L, and XL.The Medium size (45%) dominates purchases, followed by Large (27%), Small (17%), and XL (11%) *(shown in pie chart)*.

- Color Preference: Customers purchased products in 25 different colors, with Olive being the most preferred and Gold the least*(visualized in horizontal bar plot)*.

- Seasonal Purchase Trends: Purchases are evenly distributed across all four seasons, with slightly higher activity in Fall and Winter (5,867). *(shown in bar chart)*.

- The highest customer are from Montana, California, Idaho, and Illinois.(as shown in bar plot)

- Payment Methods: Six payment types were used — PayPal, Credit Card, Cash, Debit Card, Venmo, and Bank Transfer. All methods have similar usage levels *(shown in pie chart)*. The average purchase amount is also consistent across the methods, with Debit Card users spending slightly more compared toother methods *(depicted in bar plot)*.

- Shipping Type: There are six shipping options, with Free Shipping being the most popular and 2-Day Shipping the least*(visualised in bar plot)*.

- Subscription Status: About 73% customers are non-subscribers, and 27% are subscribers*(depicted in bar plot)*. Non-subscribers show stronger interest in Clothing and Accessories, while subscribers purchase fewer items overall *(shown in countplot)*.

- Customers who purchased with or without discounts show almost identical average review ratings 3.76 (No Discount) and 3.74 (Discount Applied) *(as depicted in bar plot)*.
